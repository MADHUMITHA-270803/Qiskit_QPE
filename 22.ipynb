```python
def qpe_phase_estimation(theta, counting_qubits=3):
    # Create Quantum Circuit
    qc = QuantumCircuit(counting_qubits + 1, counting_qubits)

    # Prepare the eigenstate (target qubit)
    qc.x(counting_qubits)

    # Apply Hadamard gates to counting qubits
    for q in range(counting_qubits):
        qc.h(q)
        qc.cp(2 * np.pi * theta * (2 ** q), q, counting_qubits)

    # Apply inverse QFT
    for i in range(counting_qubits // 2):
        qc.swap(i, counting_qubits - i - 1)

    for j in range(counting_qubits):
        for k in range(j):
            qc.cp(-np.pi / float(2 ** (j - k)), k, j)
        qc.h(j)

    # Measurement
    qc.measure(range(counting_qubits), range(counting_qubits))
    return qc
```

```python
# --- Task 5: Noise Simulation  ---

# âœ… Install Qiskit Aer if not already installed
!pip install qiskit qiskit-aer --quiet

from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
from qiskit_aer.noise import NoiseModel, depolarizing_error
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
import numpy as np

# --- Step 1: Define a QPE circuit function ---
def qpe_phase_estimation(theta):
    n_count = 3  # number of counting qubits
    qc = QuantumCircuit(n_count + 1, n_count)
    qc.h(range(n_count))
    qc.x(n_count)
    for q in range(n_count):
        qc.cp(2 * np.pi * theta * (2 ** q), q, n_count)
    qc.barrier()
    for j in range(n_count // 2):
        qc.swap(j, n_count - j - 1)
    for j in range(n_count):
        for k in range(j):
            qc.cp(-np.pi / float(2 ** (j - k)), k, j)
        qc.h(j)
    qc.barrier()
    qc.measure(range(n_count), range(n_count))
    return qc

# --- Step 2: Create a noise model ---
noise_model = NoiseModel()
depolar_1q = depolarizing_error(0.01, 1)
depolar_2q = depolarizing_error(0.02, 2)

noise_model.add_all_qubit_quantum_error(depolar_1q, ['h', 'x'])
noise_model.add_all_qubit_quantum_error(depolar_2q, ['cx', 'cp'])

# --- Step 3: Build and draw the circuit safely ---
theta = 0.25
qc_noise = qpe_phase_estimation(theta)

try:
    qc_noise.draw('mpl')
except Exception:
    print("âš ï¸ 'pylatexenc' not found â€” showing circuit in text mode:")
    print(qc_noise.draw('text'))

# --- Step 4: Run simulations (ideal vs noisy) ---
backend = Aer.get_backend('qasm_simulator')

ideal_result = backend.run(transpile(qc_noise, backend), shots=1024).result()
ideal_counts = ideal_result.get_counts()

noisy_result = backend.run(
    transpile(qc_noise, backend),
    shots=1024,
    noise_model=noise_model
).result()
noisy_counts = noisy_result.get_counts()

# --- Step 5: Compare results visually ---
print("\nâœ… Comparison of Ideal vs Noisy Outputs:")
plot_histogram([ideal_counts, noisy_counts], legend=['Ideal', 'Noisy'])
plt.show()
```
